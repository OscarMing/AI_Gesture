## Train Data Preprocessing ##

```python
# For train data 
import os

# dataset path
data_path ='/home/jovyan/et091-group8/GestureDataset/B'
# dataset info store path
out_filename = './list/c3d_train003.txt'
x_data_list = []
y_data_list = []
file_list = []
cls_map = {'SwipingLeft':0, 'SwipingRight':1, 'SwipingDown':2, 'SwipingUp':3, 'TurningHandClockwise':4, 'TurningHandCounterclockwise':5}
cls_map_count = {'SwipingLeft':0, 'SwipingRight':0, 'SwipingDown':0, 'SwipingUp':0, 'TurningHandClockwise':0, 'TurningHandCounterclockwise':0}
for roots, dirs, files in os.walk(data_path):
    
    for each in files:
        if each.find('checkpoint') == -1:
            xss = str(roots)+'/'
            yss = str(roots.split("/")[6])
            if xss not in x_data_list:
                x_data_list.append(xss)
                y_data_list.append(yss)
                
                if cls_map_count[yss] < 2000 and len(files) > 32:
                    video_frame_info = xss + " " + str(len(files)) + " " + str(cls_map[yss]) +'\n'
                    with open(out_filename, 'a') as out_file:
                        out_file.write(video_frame_info)
                    cls_map_count[yss]+=1
                
            else:
                pass
            
#             # Rename each frame picture
#             newname = int(each.split('.')[0])
#             os.rename(os.path.join(roots, each),os.path.join(roots, f'frm_{newname:06d}.jpg')) 
```
