## Open the labeling data ##

```python
import csv
import pandas as pd
path = './label/jester-v1-train.csv'
filename = []
filelable = []
alldata = []
with open(path, newline='') as csvfile:
    rows = csv.DictReader(csvfile, delimiter=';')
#     rows = csv.reader(csvfile, delimiter=':')

    for row in rows:
        filename.append(row['k'])
        filelable.append(row['v'])
        alldata.append([row['k'],row['v']])
```

```python
df = pd.DataFrame(alldata,columns=['k','v'])
df
```

## Filter required gestures ##
1.  Swiping Left
2.  Swiping Right
3.  Swiping Up
4.  Swiping Down
5.  Turning Hand Counterclockwise
6.  Turning Hand Clockwise

```python
# Gesture choice
lst = ['Swiping Left','Swiping Right','Swiping Up','Swiping Down','Turning Hand Counterclockwise','Turning Hand Clockwise']
tt = 0
for name, group in df.groupby('v'):
    if name in lst:
        tt+=len(group)
        print(name,':',len(group))
print(tt)
```

```python
SwipingLeft = df[df['v'] =='Swiping Left']
SwipingLeft
SwipingLeftList = list(SwipingLeft['k'].values)
len(SwipingLeftList)
# SwipingLeft.value_counts()
# SwipingLeft.v.value_counts()
SwipingLeftDict = {}
for i in range(len(SwipingLeftList)):
    dt = {SwipingLeft.iloc[i]['k']:SwipingLeft.iloc[i]['v']}
    SwipingLeftDict.update(dt)
len(SwipingLeftDict)
```

```python
SwipingRight = df[df['v'] =='Swiping Right']
SwipingRightList = list(SwipingRight['k'].values)
len(SwipingRightList)
SwipingRightDict = {}
for i in range(len(SwipingRightList)):
    dt = {SwipingRight.iloc[i]['k']:SwipingRight.iloc[i]['v']}
    SwipingRightDict.update(dt)

# SwipingRightDict
len(SwipingRightDict)
```

```python
SwipingUp = df[df['v'] =='Swiping Up']
SwipingUpList = list(SwipingUp['k'].values)
len(SwipingUpList)
SwipingUpDict = {}
for i in range(len(SwipingUpList)):
    dt = {SwipingUp.iloc[i]['k']:SwipingUp.iloc[i]['v']}
    SwipingUpDict.update(dt)

len(SwipingUpDict)
```

```python
SwipingDown = df[df['v'] =='Swiping Down']
SwipingDownList = list(SwipingDown['k'].values)
len(SwipingDownList)
SwipingDownDict = {}
for i in range(len(SwipingDownList)):
    dt = {SwipingDown.iloc[i]['k']:SwipingDown.iloc[i]['v']}
    SwipingDownDict.update(dt)

len(SwipingDownDict)
```

```python
TurningHandCounterclockwise = df[df['v'] =='Turning Hand Counterclockwise'] 
TurningHandCounterclockwiseList = list(TurningHandCounterclockwise['k'].values)
len(TurningHandCounterclockwiseList)
TurningHandCounterclockwiseDict = {}
for i in range(len(TurningHandCounterclockwiseList)):
    dt = {TurningHandCounterclockwise.iloc[i]['k']:TurningHandCounterclockwise.iloc[i]['v']}
    TurningHandCounterclockwiseDict.update(dt)

len(TurningHandCounterclockwiseDict)
```

```python
TurningHandClockwise = df[df['v'] =='Turning Hand Clockwise'] 
TurningHandClockwiseList = list(TurningHandClockwise['k'].values)
len(TurningHandClockwiseList)
TurningHandClockwiseDict = {}
for i in range(len(TurningHandClockwiseList)):
    dt = {TurningHandClockwise.iloc[i]['k']:TurningHandClockwise.iloc[i]['v']}
    TurningHandClockwiseDict.update(dt)

len(TurningHandClockwiseDict)
```

## Put all gesture into dict ##
### dict map => { gesture file name : gesture type } ###

```python
all_gesture = {}
all_gesture.update(SwipingLeftDict)
all_gesture.update(SwipingRightDict)
all_gesture.update(SwipingUpDict)
all_gesture.update(SwipingDownDict)
all_gesture.update(TurningHandCounterclockwiseDict)
all_gesture.update(TurningHandClockwiseDict)
len(all_gesture)
```

```python
import shutil
bb = all_gesture.copy()
len(bb)
```

```python
# bb = {'A01':'R','B01':'L'}
import os
import time
st = time.time()
# Origin 20BN dataset path
outerpath = '/20bn-jester-v1'
# restore path
newpath = './B/'
outerpathlist = os.listdir(outerpath)
print(len(outerpathlist))
for v in outerpathlist:
    if v in bb.keys():
        sub_files = os.path.join(outerpath, v)
#         if not os.path.exists(sub_file):
#             os.makedirs(sub_file)
        new_sub_file = os.path.join(newpath, bb[v],v)
        if not os.path.exists(new_sub_file):
            os.makedirs(new_sub_file)
            
        sub_files_list = os.listdir(sub_files)
        for f in sub_files_list:
            sub_file = os.path.join(sub_files, f)
            shutil.copy(sub_file, new_sub_file)
ed = time.time()
tt = (ed-st)/60
```
